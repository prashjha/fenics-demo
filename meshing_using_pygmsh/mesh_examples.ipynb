```python
import pyvista as pv
def plot_mesh(mfile):
    # Read the VTK file
    mesh = pv.read(mfile + '.vtk')
    
    # Create a plotter
    plotter = pv.Plotter(off_screen=True)
    plotter.add_mesh(mesh, show_edges=True)
    # plotter.view_xy()
    plotter.camera.zoom(1)
    
    # Save the plot
    plotter.screenshot(mfile + '_plot.png')
    # plotter.close()

    # plot the image with title
    plt.imshow(plotter.screenshot(mfile + '_plot.png'))
    plt.title(mfile)
    plt.show()         
            
```

# Meshing standard geomtries

```python
test_meshes = ['circle', 'ellipse', 'sphere', 'cuboid', 'ellipsoid', 'rectangle', 'hexagon', 'drum2d', 'triangle', 'polygon', 'cylindrical2d_wall', 'cylinder', 'annulus_circle', 'annulus_rectangle', 'open_rectangle', 'open_pipe']
# test_meshes = ['open_pipe']
symm_flag = 0 # 0 - both, 1 - symmetric, 2 - non-symmetric

mfile_pref = './meshes/'
os.makedirs(mfile_pref, exist_ok=True)

for tag in test_meshes:
    if tag == 'circle':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            circle_mesh_symmetric(xc = [-3., -3., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True) 
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            circle_mesh_symmetric(xc = [-3., -3., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'ellipse':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            ellipse_mesh_symmetric(xc = [-1., -1., 0.], rx = 1., ry = 0.5, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            ellipse_mesh_symmetric(xc = [-1., -1., 0.], rx = 1., ry = 0.5, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'sphere':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            sphere_mesh_symmetric(xc = [-5., -5., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            sphere_mesh_symmetric(xc = [-5., -5., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'rectangle':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            rectangle_mesh_symmetric(xc = [1., 1., 0.], Lx = 1., Ly = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)      
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            rectangle_mesh_symmetric(xc = [1., 1., 0.], Lx = 1., Ly = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)  
            plot_mesh(mfile)
    elif tag == 'hexagon':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            hexagon_mesh_symmetric(xc = [3., 3., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)  
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            hexagon_mesh_symmetric(xc = [3., 3., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'drum2d':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            drum2d_mesh_symmetric(xc = [5., 5., 0.], r = 1., width = 0.5, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)  
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            drum2d_mesh_symmetric(xc = [5., 5., 0.], r = 1., width = 0.5, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'triangle':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            triangle_mesh_symmetric(xc = [7., 7., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            triangle_mesh_symmetric(xc = [7., 7., 0.], r = 1., h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'polygon':
        theta = np.pi/6.
        R, a = 1., 0.25
        v1 = [0., 0., 0.]
        v2 = [R*np.cos(0.5*theta), -R*np.sin(0.5*theta), 0.]
        v4 = [R*np.cos(0.5*theta), R*np.sin(0.5*theta), 0.]
        v3 = [R + a, 0., 0.]
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            polygon_mesh_symmetric(points = [v1, v2, v3, v4], theta = theta, xc = [9., 9., 9.], h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            polygon_mesh_symmetric(points = [v1, v2, v3, v4], theta = theta, xc = [9., 9., 9.], h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'cylindrical2d_wall':
            cylindrical2d_wall_mesh(
                center=[0., 0., 0.],
                outer_radius=1.0,
                inner_radius=0.8,
                bar_width=0.2,
                bar_length=0.3,
                h=0.05,
                    filename='./' + tag + '_sym',
                vtk_out=True
                )
            plot_mesh(mfile)
    elif tag == 'cuboid':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            cuboid_mesh_symmetric(xc = [0., 0., 0.], Lx = 1., Ly = 0.5, Lz = 0.3, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            cuboid_mesh_symmetric(xc = [0., 0., 0.], Lx = 1., Ly = 0.5, Lz = 0.3, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'ellipsoid':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            ellipsoid_mesh_symmetric(xc = [0., 0., 0.], rx = 1., ry = 0.5, rz = 0.3, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            ellipsoid_mesh_symmetric(xc = [0., 0., 0.], rx = 1., ry = 0.5, rz = 0.3, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'cylinder':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            cylinder_mesh_symmetric(xc = [0., 0., 0.], r = 1., h = 2., mesh_size = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            cylinder_mesh_symmetric(xc = [0., 0., 0.], r = 1., h = 2., mesh_size = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'annulus_circle':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            annulus_circle_mesh_symmetric(xc = [0., 0., 0.], r_outer = 1., r_inner = 0.5, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            annulus_circle_mesh_symmetric(xc = [0., 0., 0.], r_outer = 1., r_inner = 0.5, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)  
            plot_mesh(mfile)
    elif tag == 'annulus_rectangle':
        if symm_flag == 0 or symm_flag == 1:
            mfile = mfile_pref + tag + '_sym'
            annulus_rectangle_mesh(xc = [0., 0., 0.], Lx = 1., Ly = 0.8, hole_Lx = 0.2, hole_Ly = 0.4, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = True)
            plot_mesh(mfile)
        if symm_flag == 0 or symm_flag == 2:
            mfile = mfile_pref + tag + '_non_sym'
            annulus_rectangle_mesh(xc = [0., 0., 0.], Lx = 1., Ly = 0.8, hole_Lx = 0.2, hole_Ly = 0.4, h = 0.1, filename = mfile, vtk_out = True, symmetric_mesh = False)
            plot_mesh(mfile)
    elif tag == 'open_rectangle':
        mfile = mfile_pref + tag
        open_rectangle_mesh(xc = [0., 0., 0.], Lx = 1.2, Ly = 0.8, hole_Lx = 1., hole_Ly = 0.6, h = 0.1, filename = mfile, vtk_out = True)
        plot_mesh(mfile)
    elif tag == 'open_pipe':
        mfile = mfile_pref + tag
        open_pipe_mesh(xc=[0., 0., 0.], axis=[1., 1., 0.], length=2., outer_radius=0.5, 
                wall_thickness=0.2, h=0.1, filename=mfile, vtk_out=True)
        plot_mesh(mfile)
    else:
        print(f"Mesh {tag} not found")
```
